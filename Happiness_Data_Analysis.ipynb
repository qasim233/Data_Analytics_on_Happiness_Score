```python
#Rename all columns of all datasets to have same column names
data_2015.columns = ['Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Standard Error', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia Residual']
data_2016.columns = ['Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia Residual']
data_2017.columns = ['Country', 'Happiness Rank', 'Happiness Score', 'Whisker High', 'Whisker Low', 'Economy', 'Family', 'Health', 'Freedom', 'Generosity', 'Trust', 'Dystopia Residual']
data_2018.columns = ['Happiness Rank', 'Country', 'Happiness Score', 'Economy', 'Family', 'Health', 'Freedom', 'Generosity', 'Trust']
data_2019.columns = ['Happiness Rank', 'Country', 'Happiness Score', 'Economy', 'Family', 'Health', 'Freedom', 'Generosity', 'Trust']
```

```python
#Checking for outliers in all datasets
for data in all_datasets:
    data.boxplot()
    plt.figure(figsize=(5, 5))
    plt.show()
```

```python
#remove all outliers
for data in all_datasets:
    for column in data.columns:
        if data[column].dtype != 'object':
            q1 = data[column].quantile(0.25)
            q3 = data[column].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - (1.5 * iqr)
            upper_bound = q3 + (1.5 * iqr)
            data[column] = data[column][(data[column] > lower_bound) & (data[column] < upper_bound)]
```

```python
"""Combine the datasets from 2015-2019 to construct 1 dataframe that
contains the following columns. Country, Region, Year (coming from
the respective dataset) and Happiness Score for each year as a new
column
"""

data_2015['Year'] = 2015
data_2016['Year'] = 2016
data_2017['Year'] = 2017
data_2018['Year'] = 2018
data_2019['Year'] = 2019

data = pd.concat([data_2015, data_2016, data_2017, data_2018, data_2019])
data = data[['Country', 'Year', 'Happiness Score']]
data
```

```python
"""For any 2 countries of your choice draw a line chart (x-axis contains
the years from the dataframe constructed in part g and y-axis
contains the Happiness Score values for the respective years).
Identify the trend"""

countries = data[(data['Country'] == 'India') | (data['Country'] == 'Pakistan')]
countries = countries.pivot_table(index=['Country'], columns='Year', values='Happiness Score')

countries.T.plot()
plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.title('Happiness Score of India and Pakistan')
plt.xticks(np.arange(2015, 2020))
plt.grid(True)
plt.show()
```
